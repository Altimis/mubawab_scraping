```python
def string_to_date(string):
    
    if len(string)==10:
        return datetime.strptime(string, "%d/%m/%Y")
    
    else :
        string = string[:10]
        return datetime.strptime(string, "%Y-%m-%d")
    


def keep_only_digits(string):
    if string is not None:
        
        if '²' in string:
            
        
            string = string.replace("²", "")
    
        return(''.join(filter(str.isdigit,string)))
    
    else:
        return None



def price_string_to_float(string):
    
    if isinstance(string, str):
        
        string = string.lower()
        
        if 'dh' in string:
        
            return(float(keep_only_digits(string)))
        
        elif 'eur' in string:
            
            return(float(keep_only_digits(string))*10.5)
        
        else:
            
            return(float(keep_only_digits(string))*9)
        



def get_X_Y(string, axe='X'):
    
    if string:
    
        dic_pos = eval(string)

        if axe=='X':

            return dic_pos['x']
        else:
            return dic_pos['y']
        
def get_type_from_link(link):
    
    link = link.lower()
    
    #combinaison_app = ['app','apartement','apart','apt','aprt']
    
    
        
    
    if ('app'  in link) or ('apartement' in link) or ('apart' in link) or ('apt' in link) or ('aprt' in link): #bricoul a modifi
        return 'A'
    elif 'maison' in link:
        return 'M'
    elif 'villa' in link:
        return 'V'
    elif 'riad' in link:
        return 'R'
    
    elif 'immeuble' in link:
        return 'I'
    
    elif 'studio' in link:
        return 'S'
    elif 'duplex' in link:
        return 'D'
    else:
        return 'Jesspa'
```

```python
def eval_2(string):

    if isinstance(string,str):
        return eval(string)
    else:
        return None



def columns_to_numbers(df):
    
    
    df['surface']=df['surface'].apply( lambda x : eval_2(keep_only_digits(x) )) #7adari bug hna si surface none
    
    df['pieces']=df['pieces'].apply( lambda x : eval_2(keep_only_digits(x) ))
    
    df['chambre']=df['chambre'].apply( lambda x : eval_2(keep_only_digits(x) ))
    
    
    df['bains']=df['bains'].apply( lambda x : eval_2(keep_only_digits(x) ))
    
    df['etage'] = df['etage'].apply( lambda x : eval_2(keep_only_digits(x) ))
    
    return df



def flatten_carac(df):
    L_carac = []

    for e in df.caracteristiques:
        L_carac.append(eval(e))

    merged = list(itertools.chain(*L_carac))

    unique = list(set(merged))

    unique.remove('')
    
    for var in unique:
        
        L=[]
        
        for row in L_carac:
            if var in row:
                L.append(1)
            else:
                L.append(None)
                
                
        df[var] = L
        
        
    return df
    
    
    
    
    
    
```
